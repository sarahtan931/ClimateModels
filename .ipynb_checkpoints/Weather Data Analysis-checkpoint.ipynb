<h1>Creating a climate model for Weather Data</h1>

<h3>Importing all the required libraries</h3>

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.ensemble import RandomForestRegressor
from sklearn.dummy import DummyRegressor
```

<h3>Importing the data set and using linear interpolation to reduce null values</h3>
<h3>Reading the first few lines of the data</h3>

```python
df = pd.read_csv('daily.csv', low_memory=False )

df.interpolate(method ='linear', limit_direction ='forward', inplace=True)

df.head()
```

<h3>Describing the data set </h3>

```python
df.describe()
```

<h3>Finding the features that have null values </h3>

```python
df.isnull().sum()
```

<h3>Creating a heat map to determine correlation between temperature and other features in the data set</h3>

```python
dewPoint = df['Dew Point Temp (°C)']
relHum = df['Rel Hum (%)']
windDir = df['Wind Dir (10s deg)']
windSpeed = df['Wind Spd (km/h)']
visibility = df['Visibility (km)']
stnPress = df['Stn Press (kPa)']
windChill = df['Wind Chill']

X = pd.concat([dewPoint, relHum, windDir, windSpeed, visibility, stnPress, windChill], axis=1) #features 
#X = df['Date/Time']
y = df['Temp (°C)'] #target column

#getting correlations of each feature in dataset
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
plt.title('Correlation')

g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

<h3>Creating an data frame X that has relevant features to predict y</h3>

```python
time = df['Time']
newTime = []

for index, row in time.iteritems():
     newTime.append(row[0:2])

dtime = pd.DataFrame({'col':newTime})
print(dtime)
```

```python
X1 = df['Dew Point Temp (°C)']
X2 = df['Wind Chill']
X3 = df['Visibility (km)']

year = df['Year']
month = df['Month']
day = df['Day']

#X = pd.concat([X1, X2, X3], axis=1)
X = pd.concat([year, month, day, dtime], axis=1)
y = df['Temp (°C)']

print(X.shape)
print(y.shape)
```

<h3>Splitting the data into a testing and a training set</h3>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state= 0)
```

```python
#plt.scatter(X_test['Dew Point Temp (°C)'], Y_test,  color='black', label='Dew point')
#plt.scatter(X_test['Wind Chill'], Y_test,  color='blue', label='Wind chill')
#plt.scatter(X_test['Visibility (km)'], Y_test,  color='orange', label='Visibility')
#plt.legend()
#plt.title('Correlation between features and temperature')
#plt.xlabel('Features')
#plt.ylabel('Temperature')
```

<h3>Using the Naive(copy) model to fit the data</h3>

```python
X_trainBaseline, Y_trainBaseline = X_train, Y_train
dummy_regr = DummyRegressor(strategy="mean")
dummy_regr.fit(X_trainBaseline, Y_trainBaseline)
```

<h3>Using the Random Forest Regresser model to fit the data</h3>

```python
X_train1, Y_train1 = X_train, Y_train
tree_model = RandomForestRegressor()
tree_model.fit(X_train1, Y_train1)
```

<h3>Using the linear regression model to fit the data</h3>

```python
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
```

<h3>Determining the coefficients for each feature</h3>

```python
coeff_df = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print(coeff_df)
```

<h3>Testing the models</h3>

```python
y_pred = model.predict(X_test)
y_pred1 = tree_model.predict(X_test)
y_predBaseline = dummy_regr.predict(X_test)
```

<h3>Comparing the actual outputs vs. the predicted outputs</h3>

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': y_predBaseline})
print(df.head(10))
plotBaseline = df[:10]
```

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred})
print(df.head(10))
plot = df[:10]
```

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred1})
print(df.head(10))
plot1 = df[:10]
```

<h3>Plotting the actual outputs vs. the predicted outputs</h3>

```python
plot.plot(kind='bar',figsize=(10,8))

plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title('Actual vs Predicted Temperatures with Linear Regression')
plt.ylabel('Temperature')
plt.xlabel('Datas index number')

plt.show()
```

```python
plot1.plot(kind='bar',figsize=(10,8))

plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title('Actual vs Predicted Temperatures with Random Forest Regresser')
plt.ylabel('Temperature')
plt.xlabel('Datas index number')

plt.show()
```

<h3>Determining the intercept value, the root mean square error and the r squared value to validate our model</h3>

```python
rmsd = np.sqrt(mean_squared_error(Y_test, y_predBaseline))      
r2_value = r2_score(Y_test, y_predBaseline)                     

print('Baseline: Naive(copy model)')
print("Root Mean Square Error \n", rmsd) #the lower the value the better the fit
print("R^2 Value: \n", r2_value) #the closer to 1 the better the fit
```

```python
# Root Mean Squared Deviation
rmsd = np.sqrt(mean_squared_error(Y_test, y_pred))      
r2_value = r2_score(Y_test, y_pred)   

print('Linear Regression')
print("Intercept: \n", model.intercept_)
print("Root Mean Square Error \n", rmsd) #the lower the value the better the fit
print("R^2 Value: \n", r2_value) #the closer to 1 the better the fit
```

```python
# Root Mean Squared Deviation
rmsd = np.sqrt(mean_squared_error(Y_test, y_pred1))      
r2_value = r2_score(Y_test, y_pred1)                     

print('Random Forest Regressor')
print("Root Mean Square Error \n", rmsd) #the lower the value the better the fit
print("R^2 Value: \n", r2_value) #the closer to 1 the better the fit
```
